WSe have to continue this cycle ubtil the values of m and c becomrs stable.This could be possible to do with one independent variable, bue real-world data has more than one independent variables to predict the dependent variable.So, we use gradient descent algoruthm to do this task.

## Implementing gredient descent for flat_area and sale_price, with sale_preice being the independent variable.

```python
#step -1, random initialization
def random_init(y):
    m = 0.01
    c = y.mean()
    return m,c
```

```python
#Step - 2,  generate predictions
def gen_pred(m,c,X):
    pred = []
    for i in X:
        temp = i * m + c
        pred.append(temp)
    return pred
```

```python
#step- 3, calculating cost
def cal_cost(pred,Y):
    cost = np.sum(((pred - Y) ** 2 / len(Y)))
    return cost
```

```python
#step -4(i), calculating gradients for updating values
def gradients(pred,Y,X):
    n = len(Y)
    Gm = (2/n) * np.sum((pred - Y) * X)
    Gc = (2/n) * np.sum((pred - Y))
    return Gm,Gc
```

```python
#step - 4(ii) Updating the parameters
def update_param(m_old,c_old,Gm,Gc,alpha):
    m_new = m_old - alpha*Gm
    c_new = c_old - alpha*Gc
    return m_new,c_new
```

```python
#Plotting regline after convergence  or after completing alloted number of iterations
def result(m,c,X,Y,cost,predictions,i):
    if i < 999:
        print('*******Convergence happeneed after {0} iterations***********'.format(i))
    else:
        print('*******Reg line after 1000 iterations************')
    plt.figure(figsize = (15,8))
    plt.scatter(X,Y,color = 'green',label = 'Actual Data Points')
    plt.plot(X,predictions,color = 'red',label = 'Regression line with slope {0} and intecept {1}'.format(m,c))
    plt.xlabel('Flat Area')
    plt.ylabel('Sale Price')
    plt.title('Final Regression line after implementing gradient descent')
    plt.legend(loc = 'best')
    plt.show()
```

```python
sample_2D_data.shape,sample_2D_data.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
saleprice = scaler.fit_transform(sample_2D_data['Sale_Price'].values.reshape(-1,1))
flatarea = scaler.fit_transform(sample_2D_data['Flat_Area'].values.reshape(-1,1))

max_iter = 1000
alpha = 0.01     #Learning Rate
cost_old = 0

m,c = random_init(saleprice)


for i in range(0,max_iter):
    pred = gen_pred(m,c,flatarea)
    cost_new = cal_cost(pred,saleprice)
    if(abs(cost_new - cost_old) < 10 ** (-7)):
        break
    Gm,Gc = gradients(pred,saleprice,flatarea)
    m,c = update_param(m,c,Gm,Gc,alpha)
    cost_old = cost_new
    
result(m,c,flatarea,saleprice,cost_new,pred,i)
```

```python
def remove_mc(data):
    VIF = pd.Series([variance_inflation_factor(vif_vals.values,i) for i in range(data.shape[1])] , index = data.columns)
    if(VIF.max() > 5):
        print(VIF[VIF == VIF.max()].index[0],' has been removed')
        data = data.drop(VIF[VIF == VIF.max()].index[0],axis = 1)
        return data
    else:
        print("No more Multicollinearity exists in the data")
        return data
```
